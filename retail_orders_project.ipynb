```python
# extract file from zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
# import file and handle null values
import pandas as pd
df = pd.read_csv('orders.csv', na_values=['Not Available','unknown'])
df.head(2)
df['Ship Mode'].unique()
```

```python
# rename column names
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(2)
```

```python
# derive new column discount using sale and profit
df['discount']=df['list_price']*df['discount_percent']*0.01
df.head(2)
```

```python
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
df.head(5)
```

```python
# check for the column datatype
df.dtypes
```

```python
# convert order_date into date format
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
```

```python
# remove unnecessary columns
df.drop(columns = ['list_price','cost_price', 'discount_percent'])
```
